```python
import pandas as pd
```

```python
file_path = 'data-since-april-2021.csv'

with open(file_path, mode='r', encoding='utf-8', errors='replace') as f:
    df = pd.read_csv(f, low_memory=False)
```

```python
df.info()
```

```python
df['team2-score'] = pd.to_numeric(df['team2-score'], errors='coerce')
df = df.dropna(subset=['team1-score', 'team2-score'])
df = df[df['team2-score'] != df['team1-score']]
# Criar a coluna 'match_name'
df['match_name'] = df['team1'] + ' vs ' + df['team2']

# Criar a coluna 'winner'
def determine_winner(row):
    if pd.isna(row['team2-score']):
        return 'Unknown'
    if row['team1-score'] > row['team2-score']:
        return row['team1']
    elif row['team1-score'] < row['team2-score']:
        return row['team2']
    else:
        return 'Draw'

df['winner'] = df.apply(determine_winner, axis=1)
```

```python
df_used = df[df['winner'] != 'Draw']
df_used
```

```python
df_complete = df_used[['match_name', 'map', 'player-team', 'rating', 'acs', 'kast', 'adr', 'hs', 'fk', 'fd','winner']].groupby(['match_name', 'player-team','map', 'winner']).sum().reset_index().sort_values(by='match_name')
df_complete
```

```python
print("=== ANÁLISE INICIAL DOS DADOS ===")
print(f"Shape dos dados: {df_complete.shape}")
print(f"\nColunas: {list(df_complete.columns)}")
print(f"\nPrimeiras linhas:")
print(df_complete.head())

print(f"\n=== SELECIONANDO TOP 10 TIMES ===")
team_game_counts = df_complete['player-team'].value_counts()
print(f"Total de times disponíveis: {len(team_game_counts)}")
print(f"\nJogos por time:")
print(team_game_counts.head(15))

top_10_teams = team_game_counts.head(40).index.tolist()
print(f"\n🏆 TOP 10 TIMES SELECIONADOS:")
for i, team in enumerate(top_10_teams, 1):
    games = team_game_counts[team]
    print(f"{i:2d}. {team} ({games} jogos)")

# Filtrar dataset para apenas os top 10 times
df_original = df_complete.copy()  # Backup dos dados originais
df_complete = df_complete[df_complete['player-team'].isin(top_10_teams)].copy()

print(f"\n📊 DADOS APÓS FILTRO:")
print(f"Registros originais: {len(df_original)}")
print(f"Registros filtrados: {len(df_complete)}")
print(f"Redução: {(1 - len(df_complete)/len(df_original))*100:.1f}%")
print(f"Partidas únicas: {df_complete['match_name'].nunique()}")
print(f"Mapas únicos: {list(df_complete['map'].unique())}")
```

```python
team_profiles = {}

for team in df_complete['player-team'].unique():
    team_data = df_complete[df_complete['player-team'] == team]
    
    # Estatísticas básicas
    profile = {
        # Performance geral
        'avg_rating': team_data['rating'].mean(),
        'median_rating': team_data['rating'].median(),
        'std_rating': team_data['rating'].std(),
        'max_rating': team_data['rating'].max(),
        'min_rating': team_data['rating'].min(),
        
        # Combate
        'avg_acs': team_data['acs'].mean(),
        'avg_adr': team_data['adr'].mean(),
        'avg_kast': team_data['kast'].mean(),
        'avg_hs': team_data['hs'].mean(),
        
        # First kills/deaths
        'avg_fk': team_data['fk'].mean(),
        'avg_fd': team_data['fd'].mean(),
        'fk_fd_ratio': team_data['fk'].sum() / max(team_data['fd'].sum(), 1),
        'fk_per_round': team_data['fk'].sum() / len(team_data),
        'fd_per_round': team_data['fd'].sum() / len(team_data),
        
        # Histórico de vitórias
        'total_games': len(team_data),
        'wins': len(team_data[team_data['winner'] == team]),
        'losses': len(team_data[team_data['winner'] != team]),
        'win_rate': len(team_data[team_data['winner'] == team]) / len(team_data),
        
        # Performance por mapa
        'maps_played': team_data['map'].nunique(),
        'best_map_wr': 0,  # Será calculado abaixo
        'worst_map_wr': 1,  # Será calculado abaixo
        
        # Consistência
        'acs_consistency': 1 / (team_data['acs'].std() / team_data['acs'].mean() + 0.01),
        'rating_consistency': 1 / (team_data['rating'].std() / team_data['rating'].mean() + 0.01),
        
        # Dominância
        'avg_score_diff': 0,  # Diferença média de score quando vence
    }
    
    # Performance por mapa
    map_performance = {}
    for map_name in team_data['map'].unique():
        map_data = team_data[team_data['map'] == map_name]
        map_wins = len(map_data[map_data['winner'] == team])
        map_total = len(map_data)
        map_wr = map_wins / map_total if map_total > 0 else 0
        map_performance[map_name] = {
            'games': map_total,
            'wins': map_wins,
            'win_rate': map_wr,
            'avg_rating': map_data['rating'].mean(),
            'avg_acs': map_data['acs'].mean()
        }
    
    if map_performance:
        profile['best_map_wr'] = max([stats['win_rate'] for stats in map_performance.values()])
        profile['worst_map_wr'] = min([stats['win_rate'] for stats in map_performance.values()])
    
    profile['map_performance'] = map_performance
    team_profiles[team] = profile

team_profiles
```

```python
matchups = []
teams = list(team_profiles.keys())

# Para cada partida real, criar features comparativas
for _, row in df_complete.iterrows():
    team = row['player-team']
    match_name = row['match_name']
    map_name = row['map']
    winner = row['winner']
    
    # Encontrar o oponente na mesma partida
    opponents = df_complete[(df_complete['match_name'] == match_name) & (df_complete['player-team'] != team)]['player-team'].unique()
    
    if len(opponents) > 0:
        opponent = opponents[0]
        
        # Features do time atual
        team_profile = team_profiles[team]
        opponent_profile = team_profiles[opponent]
        
        # Criar features comparativas
        matchup = {
            'team_a': team,
            'team_b': opponent,
            'map': map_name,
            'team_a_wins': 1 if winner == team else 0,
            
            # Features básicas do Team A
            'team_a_avg_rating': team_profile['avg_rating'],
            'team_a_avg_acs': team_profile['avg_acs'],
            'team_a_avg_kast': team_profile['avg_kast'],
            'team_a_avg_adr': team_profile['avg_adr'],
            'team_a_avg_hs': team_profile['avg_hs'],
            'team_a_avg_fk': team_profile['avg_fk'],
            'team_a_avg_fd': team_profile['avg_fd'],
            'team_a_win_rate': team_profile['win_rate'],
            'team_a_fk_fd_ratio': team_profile['fk_fd_ratio'],
            'team_a_total_games': team_profile['total_games'],
            'team_a_rating_consistency': team_profile['rating_consistency'],
            'team_a_acs_consistency': team_profile['acs_consistency'],
            
            # Features básicas do Team B
            'team_b_avg_rating': opponent_profile['avg_rating'],
            'team_b_avg_acs': opponent_profile['avg_acs'],
            'team_b_avg_kast': opponent_profile['avg_kast'],
            'team_b_avg_adr': opponent_profile['avg_adr'],
            'team_b_avg_hs': opponent_profile['avg_hs'],
            'team_b_avg_fk': opponent_profile['avg_fk'],
            'team_b_avg_fd': opponent_profile['avg_fd'],
            'team_b_win_rate': opponent_profile['win_rate'],
            'team_b_fk_fd_ratio': opponent_profile['fk_fd_ratio'],
            'team_b_total_games': opponent_profile['total_games'],
            'team_b_rating_consistency': opponent_profile['rating_consistency'],
            'team_b_acs_consistency': opponent_profile['acs_consistency'],
        }
        
        # Features comparativas (diferenças)
        comparison_features = ['avg_rating', 'avg_acs', 'avg_kast', 'avg_adr', 'avg_hs', 
                                'avg_fk', 'avg_fd', 'win_rate', 'fk_fd_ratio', 'total_games',
                                'rating_consistency', 'acs_consistency']
        
        for feature in comparison_features:
            team_a_val = matchup[f'team_a_{feature}']
            team_b_val = matchup[f'team_b_{feature}']
            matchup[f'diff_{feature}'] = team_a_val - team_b_val
            matchup[f'ratio_{feature}'] = team_a_val / (team_b_val + 0.001)  # Evitar divisão por zero
        
        # Features específicas do mapa
        team_map_perf = team_profile['map_performance'].get(map_name, {})
        opponent_map_perf = opponent_profile['map_performance'].get(map_name, {})
        
        matchup['team_a_map_wr'] = team_map_perf.get('win_rate', team_profile['win_rate'])
        matchup['team_b_map_wr'] = opponent_map_perf.get('win_rate', opponent_profile['win_rate'])
        matchup['team_a_map_games'] = team_map_perf.get('games', 0)
        matchup['team_b_map_games'] = opponent_map_perf.get('games', 0)
        matchup['diff_map_wr'] = matchup['team_a_map_wr'] - matchup['team_b_map_wr']
        
        matchups.append(matchup)

matchups = pd.DataFrame(matchups)
```

```python
import pandas as pd
```

```python

file_path = 'data-since-april-2021.csv'

with open(file_path, mode='r', encoding='utf-8', errors='replace') as f:
    # aqui o Python substitui caracteres inválidos por � antes de enviar
    df_data = pd.read_csv(f, low_memory=False)
```

```python
df_data
```

```python
df_data.info()
```

```python
df_data['team2-score'] = pd.to_numeric(df_data['team2-score'], errors='coerce')
df_data = df_data.dropna(subset=['team1-score', 'team2-score'])
df_data = df_data[df_data['team2-score'] != df_data['team1-score']]
# Criar a coluna 'match_name'
df_data['match_name'] = df_data['team1'] + ' vs ' + df_data['team2']

# Criar a coluna 'winner'
def determine_winner(row):
    if pd.isna(row['team2-score']):
        return 'Unknown'
    if row['team1-score'] > row['team2-score']:
        return row['team1']
    elif row['team1-score'] < row['team2-score']:
        return row['team2']
    else:
        return 'Draw'

df_data['winner'] = df_data.apply(determine_winner, axis=1)
```

```python
df_used = df_data[df_data['winner'] != 'Draw']
df_used
```

```python
df_complete = df_used[['match_name', 'map', 'player-team', 'rating', 'acs', 'kast', 'adr', 'hs', 'fk', 'fd','winner']].groupby(['match_name', 'player-team','map', 'winner']).sum().reset_index().sort_values(by='match_name')
df_complete
```

```python
df_complete.to_csv('dados_limpos.csv', index=False)
```
